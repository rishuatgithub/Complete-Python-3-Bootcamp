```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
y = x ** 2
```

```python
x
```

```python
y
```

```python
%matplotlib inline ## only required in jupyter notebook
```

```python
plt.plot(x,y)
#plt.show() # not required in jupyter
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title of the chart")
```

```python
## plot sub plots

plt.subplot(1,2,1)  ## <row>,<column>,<which block you are refering to>
plt.plot(x,y,'r')

plt.subplot(1,2,2) 
plt.plot(y,x,'b')
```

```python
### using object oriented method

fig = plt.figure()  # empty figure

## add axes to the figure

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  ## list is : (left,bottom,width,height) in relation to empty canvas/figure
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y,'r')
axes1.set_title("Larger Plot")

axes2.plot(y,x,'b')
axes2.set_title("Smaller Plot")
```

```python
## another OO approach for subplots

fig,axes = plt.subplots(nrows=1,ncols=2)  ## calls an empty canvas with 2 plot as defined in nrows and ncols. useful since you don't have to manually give the dims.

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x,'r')
axes[1].set_title("Second Plot")
```

```python
axes  ## it is nothing but an array of matplotlib axes
```

### Figure Size, DPI

```python
fig = plt.figure(figsize=(8,2),dpi=100)  ## adding fig size; dpi is optional 

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2), dpi=100)  ## adding fig size; dpi is optional 

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

### Legends

```python
fig = plt.figure(figsize=(8,5))

axes=fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='X Squared')
axes.plot(x,x**3,label='Y Squared')
axes.set_title("Plot X Squared vs Y Squared")

axes.legend(loc=0)  ## adding a legend, label is required, loc is having code from 0-10, with 0 being the best fit
```
