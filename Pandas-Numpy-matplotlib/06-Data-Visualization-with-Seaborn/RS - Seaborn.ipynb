```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Chart Examples

```python
# dist chart

sns.distplot(tips['total_bill'],kde=False)
```

```python
## joint plot

sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
## pair plot

sns.pairplot(data=tips,hue='sex',palette='coolwarm')
```

```python
## rug plot

sns.rugplot(tips['total_bill'],height=0.2)
```

## categorical data

```python
import numpy as np

sns.barplot(x='day',y='total_bill',data=tips,hue='sex',estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips,hue='day',palette='twilight_shifted_r')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```
